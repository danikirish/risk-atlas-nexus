#### This notebook explains how to uncover risks related to your usecase based on a given taxonomy.


##### Import libraries


```python
from risk_atlas_nexus.blocks.inference import (
    RITSInferenceEngine,
    WMLInferenceEngine,
    OllamaInferenceEngine,
    VLLMInferenceEngine,
)
from risk_atlas_nexus.blocks.inference.params import (
    InferenceEngineCredentials,
    RITSInferenceEngineParams,
    WMLInferenceEngineParams,
    OllamaInferenceEngineParams,
    VLLMInferenceEngineParams,
)
from risk_atlas_nexus.library import RiskAtlasNexus
```

```python
inference_engine = OllamaInferenceEngine(
    model_name_or_path="qwen2:0.5b",
    credentials=InferenceEngineCredentials(api_url="http://127.0.0.1:11434"),
    parameters=OllamaInferenceEngineParams(
        num_predict=1000, num_ctx=8192, temperature=0, repeat_penalty=1
    ),
)
```

```python
risk_atlas_nexus = RiskAtlasNexus()
```

##### Risk Identification using IBM AI Risk taxonomy


```python
def get_risks_from_usecase(usecase, taxonomies):
    available_taxonomies = [
    "ibm-risk-atlas",
    "nist-ai-rmf",
    "mit-ai-risk-repository",
    "ibm-granite-guardian",
    "credo-ucf"
]
    print(f"Identifying risks for usecase: {usecase}...")
    
    for taxonomy in taxonomies:
        print(f"\n\n=======\n Using taxonomy '{taxonomy}'")
        if taxonomy not in available_taxonomies:
            print("Taxonomy not part of available taxonomies! Skipping...")
            continue
        
        risks = risk_atlas_nexus.identify_risks_from_usecases(
        usecases=[usecase],
        inference_engine=inference_engine,
        taxonomy=taxonomy,
        max_risk=5,
    )
    
        print(f"\nIdentified {len(risks)} risks")
        for i, risk in enumerate(risks):
            risk = risk[0]
            print(f"\n{i+1}. {risk.name}")
            print(f"Description: {risk.description}")
            print(f"Concern (reason): {risk.concern}")
            print(f"Source: {risk.url}")
        
```

```python
taxonomies = [
    "ibm-risk-atlas",
    "nist-ai-rmf",
    "mit-ai-risk-repository",
    "ibm-granite-guardian",
    "credo-ucf"
]
    
```

```python
get_risks_from_usecase("Generate personalized, relevant responses, recommendations, and summaries of claims for customers to support agents to enhance their interactions with customers.", ["ibm-risk-atlas", "nist-ai-rmf"])
```
